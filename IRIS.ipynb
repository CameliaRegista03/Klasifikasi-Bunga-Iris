### **IMPORT LIBRARY**

```python
import pandas as pd
from sklearn.datasets import load_iris

#Baca Dataset
iris = pd.read_csv("Iris.csv")
```

```python
#Tampilkan data teratas iris

print("Tampilkan data teratas dari dataset iris :")
iris.head()
```

```python
#Menghapus kolom "Id" yang tidak berguna menggunkan fungsi drop()
iris.drop('Id', axis=1,inplace=True)
```

```python
#Pisahkan atribut
x = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = iris['Species']
```

```python
#Melatih data dengan fungsi fit()
from sklearn.tree import DecisionTreeClassifier

#buat model decision tree
tree_model = DecisionTreeClassifier()

#Melatih data
tree_model.fit(x, y)
```

```python
#Mencoba model 
tree_model.predict([[5.7, 7.8, 4.7, 8.7]]) #urutan atribut (panjang sepal, lebar sepal, panjang petal, lebar petal) model ini menghasilkan prediksi iris versicolor
```

```python
#Visualisasi dengan library graphviz

from sklearn.tree import export_graphviz
export_graphviz(
    tree_model,
    out_file = "iris_tree.dot",
    feature_names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],
    class_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
    rounded = True,
    filled =True
)
```
